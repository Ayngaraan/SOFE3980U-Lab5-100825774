##5. Set Up Great Expectations Context and Data Source

```python
# 1. Initialize the Context
context = gx.get_context()

# 2. Set up a Pandas Data Source
data_source = context.data_sources.add_pandas("labels_source")
```

##6. Define and Create a Data Batch

```python
# 1. Define the asset (your dataset)
data_asset = data_source.add_dataframe_asset(name="labels_asset")

# 2. Define how you want to batch it (we want the whole DataFrame)
batch_definition = data_asset.add_batch_definition_whole_dataframe("labels_batch")

# 3. Create the actual batch using your 'df'
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

### Expectation 1

```python
expectation_1 = gx.expectations.ExpectColumnValuesToNotBeNull(
    column="Timestamp"
)
```

### Validate Data Against Expectation 1
